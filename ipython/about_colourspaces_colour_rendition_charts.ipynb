```python
import numpy as np
import pylab
from pprint import pprint

import colour
from colour.characterisation.dataset.colour_checkers.spds import (
    COLOURCHECKER_INDEXES_TO_NAMES_MAPPING)
from colour.plotting import *
from colour.utilities.verbose import message_box

pylab.rcParams['figure.figsize'] = 28, 14

COLOUR_CHECKER = colour.COLOURCHECKERS_SPDS['ColorChecker N Ohta']
REFERENCE_SPD = COLOUR_CHECKER['black 2 (1.5 D)']
SHAPE = REFERENCE_SPD.shape

# Tested colourspaces.
C_TS = (colour.sRGB_COLOURSPACE,
        colour.ACES_RGB_COLOURSPACE,
        colour.REC_2020_COLOURSPACE,
        colour.ALEXA_WIDE_GAMUT_RGB_COLOURSPACE,
        colour.S_LOG_COLOURSPACE,
        colour.DCI_P3_COLOURSPACE)

# TODO: Merge into *colour*.
COLOURSPACES_ILLUMINANTS = {'ACES RGB': 'D60',
                            'ALEXA Wide Gamut RGB': 'D65',
                            'DCI-P3': 'D65',
                            'Rec. 2020': 'D65',
                            'S-Log': 'D65',
                            'sRGB': 'D65'}

# CIE Standard Illuminant E aliases.
E = colour.ILLUMINANTS['cie_2_1931']['E']
E_SPD = colour.ILLUMINANTS_RELATIVE_SPDS['E'].clone().align(SHAPE)


def normal_distribution_function(x, mu, sigma):
    return (np.exp(-np.power(x - mu, 2.) /
                   (2 * np.power(sigma, 2.))))


# TODO: Merge into *colour*.
def RGB_colour_checker_plot(RGB, chart_title='Custom', **kwargs):
    colour_parameters = []
    for index, sample in sorted(
            COLOURCHECKER_INDEXES_TO_NAMES_MAPPING.items()):
        colour_parameters.append(
            colour_parameter(sample, np.clip(np.ravel(RGB[index - 1]), 0, 1)))

    background_colour = '0.1'
    matplotlib.pyplot.gca().patch.set_facecolor(background_colour)

    width = height = 1.0
    spacing = 0.25
    across = 6

    settings = {'standalone': False,
                'width': width,
                'height': height,
                'spacing': spacing,
                'across': across,
                'text_size': 8,
                'margins': [-0.125, 0.125, -0.5, 0.125]}
    settings.update(kwargs)

    multi_colour_plot(colour_parameters, **settings)

    text_x = width * (across / 2) + (across * (spacing / 2)) - spacing / 2
    text_y = -(len(colour_parameters) / across + spacing / 2)

    pylab.text(text_x,
               text_y,
               '{0} - Colour Rendition Chart'.format(chart_title),
               color='0.95',
               clip_on=True,
               ha='center')

    settings.update({'facecolor': background_colour,
                     'edgecolor': None,
                     'standalone': True})

    bounding_box(**settings)
    aspect(**settings)

    return display(**settings)


def get_colour_checker_sdps(colour_checker=COLOUR_CHECKER):
    spds = []
    for index, sample in sorted(
            COLOURCHECKER_INDEXES_TO_NAMES_MAPPING.items()):
        spds.append(colour_checker[sample].clone())
    return spds


def spectral_to_RGB(spds, c_r):
    RGB = []
    for spd in spds:
        # Converting each spd to *RGB* reference colourspace.
        XYZ = colour.spectral_to_XYZ(
            spd,
            illuminant=colour.ILLUMINANTS_RELATIVE_SPDS[
                COLOURSPACES_ILLUMINANTS[c_r.name]]) / 100
        RGB.append(colour.XYZ_to_RGB(XYZ,
                                     c_r.whitepoint,
                                     c_r.whitepoint,
                                     c_r.XYZ_to_RGB_matrix))
    return RGB


def filter_spds_to_RGB(spds, spd_f, c_t, c_r):
    # *RGB* test colourspace *c_t* aliases.
    c_t_w_spd = colour.ILLUMINANTS_RELATIVE_SPDS[
        COLOURSPACES_ILLUMINANTS[c_t.name]]
    c_t_w = c_t.whitepoint
    c_t_p = c_t.primaries
    c_t_XYZ_to_RGB = c_t.XYZ_to_RGB_matrix
    c_t_RGB_to_XYZ = c_t.RGB_to_XYZ_matrix

    # *RGB* reference colourspace *c_r* aliases.
    c_r_w = c_r.whitepoint
    c_r_p = c_r.primaries
    c_r_XYZ_to_RGB = c_r.XYZ_to_RGB_matrix
    c_r_RGB_to_XYZ = c_r.RGB_to_XYZ_matrix

    # Converting spds to *RGB* test colourspace *c_t*.
    RGB = spectral_to_RGB(spds, c_t)

    # Converting filter spd *RGB* test colourspace *c_t*.
    RGB_f = spectral_to_RGB([spd_f], c_t)[0]

    # Applying filter.
    RGB *= RGB_f

    # Computing transformation matrix from *RGB* test colourspace *c_t*
    # to *RGB* reference colourspace *c_r*.
    cat = colour.chromatic_adaptation_matrix(colour.xy_to_XYZ(c_t_w),
                                             colour.xy_to_XYZ(c_r_w))
    RGB_t_to_RGB_r_matrix = np.dot(c_r_XYZ_to_RGB,
                                   np.dot(cat, c_t_RGB_to_XYZ))

    # Converting to *RGB* reference colourspace *c_r*.
    RGB = [np.dot(RGB_t_to_RGB_r_matrix, x) for x in RGB]

    return RGB


def spectral_vs_colourspaces(filterers,
                             c_r,
                             c_ts=C_TS,
                             colour_checker=COLOUR_CHECKER):
    results = {}
    for filterer in filterers:
        spds = get_colour_checker_sdps(colour_checker)
        for spd in spds:
            spd *= filterer

        # Computing the spectral reference *RGB* values under
        # *RGB* reference colourspace *c_r*.
        RGB_r = spectral_to_RGB(spds, c_r)
        results[filterer.name] = {'spectral': RGB_r}

        for c_t in c_ts:
            # Computing the test reference *RGB* values under
            # *RGB* test colourspace *c_t*.
            RGB_c = filter_spds_to_RGB(
                get_colour_checker_sdps(colour_checker),
                filterer,
                c_t,
                c_r)

            results[filterer.name][c_t.name] = RGB_c
    return results


def RGB_to_Lab(RGB, c_t, c_r):
    Lab = [colour.XYZ_to_Lab(
        colour.RGB_to_XYZ(x,
                          c_r.whitepoint,
                          c_t.whitepoint,
                          c_t.RGB_to_XYZ_matrix),
        c_t.whitepoint) for x in RGB]
    return Lab


def per_filter_delta_E(results, c_r, c_ts=C_TS):
    colour_differences = {}
    for filterer, result in results.items():
        colour_differences[filterer] = {}

        Lab_r = RGB_to_Lab(result['spectral'], c_r, c_r)
        for c_t in c_ts:
            Lab = RGB_to_Lab(result[c_t.name], c_t, c_r)
            delta_E = [colour.delta_E_CIE_2000(Lab_r[i], Lab[i])
                       for i in range(len(Lab_r))]
            colour_differences[filterer][c_t.name] = np.average(
                delta_E)
    return colour_differences


def overall_delta_E(filter_colour_differences, c_ts=C_TS):
    overall_colour_differences = {}
    for c_t in c_ts:
        overall_colour_differences[c_t.name] = []
        for delta_E in filter_colour_differences.values():
            overall_colour_differences[c_t.name].append(
                delta_E[c_t.name])
        overall_colour_differences[c_t.name] = np.average(
            overall_colour_differences[c_t.name])
    return overall_colour_differences


def print_delta_E_results(colour_differences):
    for colourspace, delta_E in sorted(overall_delta_E(
            colour_differences).items(), key=lambda x: x[1]):
        print('{0}: {1}'.format(colourspace, delta_E))
```

## Broad-band Filters - Design

```python
def filters_CIE_1931_chromaticity_diagram_plot(
        filterers,
        illuminant=colour.ILLUMINANTS_RELATIVE_SPDS['D65'],
        **kwargs):
    CIE_1931_chromaticity_diagram_plot(standalone=False,
                                       **kwargs)
    for filterer in filterers:
        XYZ = colour.spectral_to_XYZ(filterer, illuminant=illuminant) / 100
        xy = colour.XYZ_to_xy(XYZ)

        pylab.plot(xy[0], xy[1], 'o', color='white')

        if filterer.name is not None:
            pylab.annotate(filterer.name,
                           xy=xy,
                           xytext=(50, 30),
                           textcoords='offset points',
                           arrowprops=dict(arrowstyle='->',
                                           connectionstyle='arc3, rad=0.2'))

    display(standalone=True)


def filter_spd(mu, sigma, shape):
    samples_count = 100
    samples = np.linspace(0, 1, samples_count)

    filter_samples = normal_distribution_function(samples, mu, sigma)
    filter_data = dict(zip(
        np.linspace(shape.start, shape.end, samples_count), filter_samples))
    filter_spd = colour.SpectralPowerDistribution(
        '{0} - {1} Filter'.format(np.around(mu, 3), np.around(sigma, 3)),
        filter_data)

    return filter_spd.align(shape).normalise()

# Defining broad-band filters.
broadband_filters = [filter_spd(x, 0.125, SHAPE) for x in
                     np.linspace(0, 1, 10)]

multi_spd_plot(broadband_filters, use_spds_colours=True)

filters_CIE_1931_chromaticity_diagram_plot(
    broadband_filters,
    title='Broad-band Filters - Chromaticity Coordinates')
```

## Narrow-band Filters - Design

```python
# Defining narrow-band filters.
narrowband_filters = [filter_spd(x, 0.05, SHAPE) for x in np.linspace(0, 1, 10)]

multi_spd_plot(narrowband_filters, use_spds_colours=True)

filters_CIE_1931_chromaticity_diagram_plot(
    narrowband_filters,
    title='Narrow-band Filters - Chromaticity Coordinates')
```

## Wide-band Filters

```python
# Defining wide-band filters.
wideband_filters = [filter_spd(x, 0.25, SHAPE) for x in np.linspace(0, 1, 10)]

multi_spd_plot(wideband_filters, use_spds_colours=True)

filters_CIE_1931_chromaticity_diagram_plot(
    wideband_filters,
    title='Wide-band Filters - Chromaticity Coordinates')
```

## Custom Filters - Design

```python
def label_custom_filters(custom_filters, label=None):
    [setattr(custom_filter, 'name', '{0} {1}'.format(label, i) if label else None ) 
     for i, custom_filter in enumerate(custom_filters)]

surface=5,
spacing=0.085
cmfs = colour.CMFS['CIE 1931 2 Degree Standard Observer']

XYZ = [value for key, value in cmfs]

path = matplotlib.path.Path([colour.XYZ_to_xy(x) for x in XYZ])
custom_filters = []
for i in np.arange(0, 1, spacing):
    for j in np.arange(0, 1, spacing):
        if path.contains_path(matplotlib.path.Path([[i, j], [i, j]])):
            XYZ = colour.xy_to_XYZ((i, j))
            RGB = colour.XYZ_to_RGB(XYZ,
                                     E,
                                     E,
                                     colour.sRGB_COLOURSPACE.XYZ_to_RGB_matrix)
            custom_filters.append(RGB)
 
custom_filters = [colour.RGB_to_spectral_smits1999(custom_filter).align(SHAPE).normalise() 
                  for custom_filter in custom_filters]

label_custom_filters(custom_filters, 'Custom')

multi_spd_plot(custom_filters, use_spds_colours=True, legend=False)

label_custom_filters(custom_filters, None)

filters_CIE_1931_chromaticity_diagram_plot(
    custom_filters,
    colour.ILLUMINANTS_RELATIVE_SPDS['E'],
    title='Custom Filters - Chromaticity Coordinates')

label_custom_filters(custom_filters, 'Custom')
```

## Filters - Aggregated $\Delta_{E_{ab}}$ CIE 2000 Colour Difference

```python
all_filters = (broadband_filters +
               narrowband_filters +
               wideband_filters +
               custom_filters)

colourspace_r_colour_differences = {}
for c_r in C_TS:
    custom_filters_results = spectral_vs_colourspaces(custom_filters, c_r)

    colourspace_r_colour_differences[c_r.name] = overall_delta_E(
        per_filter_delta_E(custom_filters_results, c_r))

message_box('Filters - Delta E')
print_delta_E_results(colourspace_r_colour_differences)
```

## CIE Standard Illuminant D Series D65 - $\Delta_{E_{ab}}$ CIE 2000 Colour Difference

```python
D65 = colour.ILLUMINANTS_RELATIVE_SPDS['D65'].clone().align(SHAPE)

colourspace_r_colour_differences = {}
for c_r in C_TS:
    D65_results = spectral_vs_colourspaces([D65], c_r)

    colourspace_r_colour_differences[c_r.name] = overall_delta_E(
        per_filter_delta_E(D65_results, c_r))

message_box('CIE Standard Illuminant D Series D65 - Delta E')
print_delta_E_results(colourspace_r_colour_differences)
```
