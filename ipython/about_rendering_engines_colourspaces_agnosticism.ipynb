# About Rendering Engines Colourspaces Agnosticism

It has pretty much always been assumed by a lot of people, including myself, that rendering engines are colourspaces agnostic, and whatever your primaries are, it should not make any differences.

## Introduction

An interesting thread with some VFX industry veterans about the [Academy Color Encoding System](http://www.oscars.org/science-technology/council/projects/aces.html) has been ignited by Steve Agland. He was describing issues while rendering in ACES RGB colourspace where the rendering equation could generate some very saturated colours, colour clipping and loss of details.

As we were discussing, [Zap 'Master' Anderson](http://mentalraytips.blogspot.fr/) pointed that he was assumming primaries wouldn't matter but that it was wrong.

I initially thought that he was incorrect, this would imply that we may have been doing rendering falsely for many years. I have never really think about it because it was kind of a postulate.

## Colourspaces Agnosticism

We will demonstrate the issue we were speaking about, notice that it doesn't actually just only concern rendering.

Starting with some references colours in *CIE XYZ* colourspace (*dark skin* and *green*) from *ColorChecker 2005*. You can use one colour to showcase the issue but we will use two because in a renderer you would most likely have interactions of multiple colours together.

We can for instance assume that *dark skin* is a texture and *green* a light.

We convert them to *sRGB* and *Adobe RGB 1998* colourspaces, we have chosen those 2 colourspaces because they have the same whitepoint and thus discount chromatic adaptation, essentially removing a problem from the equation in order to make things simpler to understand:

```python
import numpy as np
import colour
from colour.utilities.verbose import message_box

name, data, illuminant = colour.COLOURCHECKERS['ColorChecker 2005']

sRGB_w = colour.sRGB_COLOURSPACE.whitepoint
sRGB_XYZ_to_RGB = colour.sRGB_COLOURSPACE.to_RGB
sRGB_RGB_to_XYZ = colour.sRGB_COLOURSPACE.to_XYZ
adobe98_w = colour.ADOBE_RGB_1998_COLOURSPACE.whitepoint
adobe98_XYZ_to_RGB = colour.ADOBE_RGB_1998_COLOURSPACE.to_RGB
adobe98_RGB_to_XYZ = colour.ADOBE_RGB_1998_COLOURSPACE.to_XYZ

# Preparing *dark skin* colour in various colourspaces.
index, name, x, y, Y = data[0]
XYZ_r1 = colour.xyY_to_XYZ((x, y, Y))

# *sRGB* colourspace 0-255 values with OETF applied.
sRGB_rd1 = colour.XYZ_to_sRGB(XYZ_r1, illuminant)

# *sRGB* colourspace linear values.
sRGB_r1 = colour.XYZ_to_RGB(XYZ_r1,
                            illuminant,
                            sRGB_w,
                            sRGB_XYZ_to_RGB)

# *Adobe RGB 1998* colourspace linear values.
adobe98_r1 = colour.XYZ_to_RGB(XYZ_r1,
                               illuminant,
                               adobe98_w,
                               adobe98_XYZ_to_RGB)

message_box(('Reference "dark skin" "CIE XYZ" colourspace tristimulus '
             'values:\n'
             '\t{0}\n'
             '\n"sRGB" colourspace values (OETF):\n'
             '\n\t{1}\n'
             '\n"sRGB" and "Adobe RGB 1998" colourspaces (Linear):\n'
             '\tsRGB: {2}\n\tAdobe RGB 1998: {3}').format(
    XYZ_r1,
    np.around(sRGB_rd1 * 255),
    sRGB_r1,
    adobe98_r1))

# Preparing *green* colour in various colourspaces.
index, name, x, y, Y = data[13]
XYZ_r2 = colour.xyY_to_XYZ((x, y, Y))

# *sRGB* colourspace 0-255 values with OETF applied.
sRGB_rd2 = colour.XYZ_to_sRGB(XYZ_r2, illuminant)

# *sRGB* colourspace linear values.
sRGB_r2 = colour.XYZ_to_RGB(XYZ_r2,
                            illuminant,
                            sRGB_w,
                            sRGB_XYZ_to_RGB)

# *Adobe RGB 1998* colourspace linear values.
adobe98_r2 = colour.XYZ_to_RGB(XYZ_r2,
                               illuminant,
                               adobe98_w,
                               adobe98_XYZ_to_RGB)

message_box(('Reference "green" "CIE XYZ" colourspace tristimulus '
             'values:\n'
             '\t{0}\n'
             '\n"sRGB" colourspace values (OETF):\n'
             '\t{1}\n'
             '\n"sRGB" and "Adobe RGB 1998" colourspaces (Linear):\n'
             '\tsRGB: {2}\n\tAdobe RGB 1998: {3}').format(
    XYZ_r2,
    np.around(sRGB_rd2 * 255),
    sRGB_r2,
    adobe98_r2))
```

So far so good, everything is setup and we are ready to do some maths but first, as a sanity check we ensure that *dark skin* properly converts back to *CIE XYZ* tristimulus values:

```python
XYZ_sRGB1 = colour.RGB_to_XYZ(sRGB_r1,
                              sRGB_w,
                              illuminant,
                              sRGB_RGB_to_XYZ)

XYZ_adobe981 = colour.RGB_to_XYZ(adobe98_r1,
                                 adobe98_w,
                                 illuminant,
                                 adobe98_RGB_to_XYZ)

message_box(('Converting back "dark skin" "CIE XYZ" colourspace '
             'tristimulus values from "sRGB" and "Adobe RGB 1998" '
             'colourspaces:\n'
             '\tFrom sRGB: {0}\n\tFrom Adobe RGB 1998: {1}\n'
             '\nEverything looks fine!').format(
    XYZ_sRGB1, XYZ_adobe981))
```

Now some maths, we multiply *dark skin* colour by *green* multiple times and scales it a little bit to have decent values:

$$
\begin{equation}
\begin{aligned}
sRGB_m&= sRGB_{r1}\cdot sRGB_{r2}\cdot sRGB_{r2}\cdot sRGB_{r2}\cdot sRGB_k\\
adobe98_m&= adobe98_{r1}\cdot adobe98_{r2}\cdot adobe98_{r2}\cdot adobe98_{r2}\cdot adobe98_k
\end{aligned}
\end{equation}
$$

Where

$$
\begin{equation}
\begin{aligned}
sRGB_{r1}&=\ 'dark\ skin'\\
sRGB_{r2}&=\ 'green'\\
adobe98_{r1}&=\ 'dark\ skin'\\
adobe98_{r2}&=\ 'green'
\end{aligned}
\end{equation}
$$

With $sRGB_k$ and $adobe98_k$ being arbitrary scaling factors equivalent in each colourspace.

We then convert everything back to "CIE XYZ":

```python
k = np.array([500, 500, 500])

sRGB_k = colour.XYZ_to_RGB(k,
                           illuminant,
                           sRGB_w,
                           sRGB_XYZ_to_RGB)
adobe98_k = colour.XYZ_to_RGB(k,
                              illuminant,
                              adobe98_w,
                              adobe98_XYZ_to_RGB)

sRGB_m = sRGB_r1 * sRGB_r2 * sRGB_r2 * sRGB_r2 * sRGB_k
adobe98_m = adobe98_r1 * adobe98_r2 * adobe98_r2 * adobe98_r2 * adobe98_k

XYZ_sRGB_m1 = colour.RGB_to_XYZ(sRGB_m,
                                sRGB_w,
                                illuminant,
                                sRGB_RGB_to_XYZ)

XYZ_adobe98_m1 = colour.RGB_to_XYZ(adobe98_m,
                                   adobe98_w,
                                   illuminant,
                                   adobe98_RGB_to_XYZ)

message_box(('Multiplying "dark skin" with "green" and converting back to '
             '"CIE XYZ" colourspace tristimulus values from "sRGB" and '
             '"Adobe RGB 1998" colourspaces:\n'
             '\tFrom sRGB: {0}\n\tFrom Adobe RGB 1998: {1}\n'
             '\nHouston? We have a problem!').format(
    XYZ_sRGB_m1, XYZ_adobe98_m1))
```

Notice how the *CIE XYZ* colourspace tristimulus values are far from being equal?

This implies that if renderings were to be done using two different colourspaces, they would not converge to the same image, secondary bounces would be different, etc...

If you were to see the two *dark_skin* colours lit by the *green* light they would look something like that:

```python
from colour.plotting import *

sRGB_m1 = colour.XYZ_to_sRGB(XYZ_sRGB_m1)
adobe98_m1 = colour.XYZ_to_sRGB(XYZ_adobe98_m1)

nrm = np.max((sRGB_m1, adobe98_m1))
sRGB_m1 = np.clip(sRGB_m1 / nrm, 0, 1)
adobe98_m1 = np.clip(adobe98_m1 / nrm, 0, 1)

multi_colour_plot([colour_parameter('From sRGB', sRGB_m1), colour_parameter('From Adobe RGB 1998', adobe98_m1)], text_size=32)
```

We will plot this on the chromaticity diagram:

```python
colourspaces_CIE_1931_chromaticity_diagram_plot(
    ['sRGB', 'Adobe RGB 1998', 'Pointer Gamut'],
standalone=False,
title='"dark skin" Colour Computation')

for name, XYZ in (('From sRGB', XYZ_sRGB_m1),
                  ('From Adobe RGB 1998', XYZ_adobe98_m1)):
    
    xy = colour.XYZ_to_xy(XYZ)

    pylab.plot(xy[0], xy[1], 'o', color='white')
    pylab.annotate(name,
                   xy=xy,
                   xytext=(50, 30),
                   textcoords='offset points',
                   arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=0.2'))

display(standalone=True)
```

We can calculate the colour difference $\Delta_E$ to mathematically quantify the difference:

```python
Lab1 = colour.XYZ_to_Lab(XYZ_sRGB_m1, illuminant)
Lab2 = colour.XYZ_to_Lab(XYZ_adobe98_m1, illuminant)

print(colour.delta_E_CMC(Lab1, Lab2, l=1))
```

In this case, 8.1 is way above the just noticable threshold ($JND=2.3$), and to make matter worse remember the colours have been clipped which lessen the difference.

## With ACES RGB...

We will do the same computations but against *ACES RGB* colourspace and the whole colour rendition chart this time:

```python
from pprint import pprint

aces_w = colour.ACES_RGB_COLOURSPACE.whitepoint
aces_XYZ_to_RGB = colour.ACES_RGB_COLOURSPACE.to_RGB
aces_RGB_to_XYZ = colour.ACES_RGB_COLOURSPACE.to_XYZ

aces_r2 = colour.XYZ_to_RGB(XYZ_r2,
                            illuminant,
                            aces_w,
                            aces_XYZ_to_RGB)
aces_k = colour.XYZ_to_RGB(k,
                           illuminant,
                           aces_w,
                           aces_XYZ_to_RGB)

XYZs_m = []
for index, name, x, y, Y in data:
    xyY = np.array([x, y, Y])
    sRGB_r1 = colour.XYZ_to_RGB(
        colour.xyY_to_XYZ(xyY),
        illuminant,
        sRGB_w,
        sRGB_XYZ_to_RGB)
    sRGB_m = sRGB_r1 * sRGB_r2 * sRGB_r2 * sRGB_r2 * sRGB_k

    aces_r1 = colour.XYZ_to_RGB(
        colour.xyY_to_XYZ(xyY),
        illuminant,
        aces_w,
        aces_XYZ_to_RGB)
    aces_m = aces_r1 * aces_r2 * aces_r2 * aces_r2 * aces_k
    
    XYZ_sRGB_m1 = colour.RGB_to_XYZ(sRGB_m,
                                    sRGB_w,
                                    illuminant,
                                    sRGB_RGB_to_XYZ)
    XYZ_aces_m1 = colour.RGB_to_XYZ(aces_m,
                                    aces_w,
                                    illuminant,
                                    aces_RGB_to_XYZ)
    XYZs_m.append((XYZ_sRGB_m1, XYZ_aces_m1))

pprint(XYZs_m)
```

We can then compute the colour difference $\Delta_E$ for each sample:

```python
deltas_E = []
for i, (XYZ1, XYZ2) in enumerate(XYZs_m):
    Lab1 = colour.XYZ_to_Lab(XYZ1, illuminant)
    Lab2 = colour.XYZ_to_Lab(XYZ2, illuminant)
    deltas_E.append((data[i][1], colour.delta_E_CMC(Lab1, Lab2, l=1)))

pprint(deltas_E)
```

As you can see, depending the samples colour difference $\Delta_E$ can reach dramatic values. The overall colour difference on the whole colour rendition chart is as follows:

```python
np.average([delta_E[1] for delta_E in deltas_E])
```

This is more than 10 times the $JND$!!!

## With Rec. 2020...

Again, the same computations but this time against *Rec. 2020* colourspace:

```python
from pprint import pprint

rec2020_w = colour.REC_2020_COLOURSPACE.whitepoint
rec2020_XYZ_to_RGB = colour.REC_2020_COLOURSPACE.to_RGB
rec2020_RGB_to_XYZ = colour.REC_2020_COLOURSPACE.to_XYZ

rec2020_r2 = colour.XYZ_to_RGB(XYZ_r2,
                               illuminant,
                               rec2020_w,
                               rec2020_XYZ_to_RGB)
rec2020_k = colour.XYZ_to_RGB(k,
                              illuminant,
                              aces_w,
                              aces_XYZ_to_RGB)
XYZs_m = []
for index, name, x, y, Y in data:
    xyY = np.array([x, y, Y])
    sRGB_r1 = colour.XYZ_to_RGB(
        colour.xyY_to_XYZ(xyY),
        illuminant,
        sRGB_w,
        sRGB_XYZ_to_RGB)
    sRGB_m = sRGB_r1 * sRGB_r2 * sRGB_r2 * sRGB_r2 * sRGB_k

    rec2020_r1 = colour.XYZ_to_RGB(
        colour.xyY_to_XYZ(xyY),
        illuminant,
        rec2020_w,
        rec2020_XYZ_to_RGB)
    rec2020_m = rec2020_r1 * rec2020_r2 * rec2020_r2 * rec2020_r2 * rec2020_k

    XYZ_sRGB_m1 = colour.RGB_to_XYZ(sRGB_m,
                                    sRGB_w,
                                    illuminant,
                                    sRGB_RGB_to_XYZ)
    XYZ_rec2020_m1 = colour.RGB_to_XYZ(rec2020_m,
                                       rec2020_w,
                                       illuminant,
                                       rec2020_RGB_to_XYZ)
    XYZs_m.append((XYZ_sRGB_m1, XYZ_rec2020_m1))

pprint(XYZs_m)
```

We again compute the colour difference $\Delta_E$:

```python
deltas_E = []
for i, (XYZ1, XYZ2) in enumerate(XYZs_m):
    Lab1 = colour.XYZ_to_Lab(XYZ1, illuminant)
    Lab2 = colour.XYZ_to_Lab(XYZ2, illuminant)
    deltas_E.append((data[i][1], colour.delta_E_CMC(Lab1, Lab2, l=1)))

pprint(deltas_E)
```

Finally, we compute the average colour difference:

```python
np.average(np.average([delta_E[1] for delta_E in deltas_E]))
```

## Analytical Explanation

Converting from a colourspace $RGB_1$ to another colourspace $RGB_2$ is done by applying a serie of linear transformations (matrices) as follows:

$$
\begin{equation}
\begin{vmatrix}
R_2\\
G_2\\
B_2
\end{vmatrix}=
\textbf{M}_{RGB_2}\cdot \textbf{M}_{XYZ_1}
\begin{vmatrix}
R_1\\
G_1\\
B_1
\end{vmatrix}
\end{equation}
$$

where $\textbf{M}_{RGB_2}$ is the $3x3$ matrix transforming from *CIE XYZ* colourspace to $RGB_2$ colourspace and $\textbf{M}_{XYZ_1}$ is the $3x3$ matrix transforming from $RGB_1$ colourspace to *CIE XYZ* colourspace.

> Note: We have discounted chromatic adaptation to makes things easier.

Here are such $\textbf{M}_{XYZ_1}$ matrices for respectively *sRGB* and *ACES RGB* colourspaces:

$$
\begin{equation}
\textbf{M}_{XYZ_{sRGB}}=
\begin{vmatrix}
0.41238656&0.35759149&0.18045049\\
0.21263682&0.71518298&0.0721802\\
0.01933062&0.11919716&0.95037259\\
\end{vmatrix}\\
\textbf{M}_{XYZ_{ACES}}=
\begin{vmatrix}
0.9525524&0&0.00009368\\
0.34396645&0.7281661&-0.07213255\\
0&0&1.00882518\\
\end{vmatrix}\\
\end{equation}
$$

Matrices are really just a system of multiple linear equations, thus we can express the $\textbf{M}_{XYZ_{sRGB}}$ as follows:

$$
\begin{equation}
\begin{aligned}
X&=\ 0.41238656R+0.35759149G+0.18045049B\\
Y&=\ 0.21263682R+0.71518298G+0.0721802B\\
Z&=\ 0.01933062R+0.11919716G+0.95037259B\\
\end{aligned}
\end{equation}
$$

We will simplify the problem again by just considering the first equation of each system, the first row of each matrix:

$$
\begin{equation}
\begin{aligned}
X_{sRGB}&=\ 0.41238656R+0.35759149G+0.18045049B\\
X_{ACES}&=\ 0.9525524R+0.0G+0.00009368B\\
\end{aligned}
\end{equation}
$$

We plot the functions by using a uniformly spaced incrementing $k$ variable in place of $RGB$:

```python
k = np.linspace(0, 1, 10)
X_sRGB = lambda x: 0.41238656 * x + 0.35759149 * x + 0.18045049 * x
X_ACES = lambda x: 0.9525524 * x + 0.00009368 * x

pylab.plot(k, tuple(map(X_sRGB, k)), 'o-', color='red', label='X - sRGB')
pylab.plot(k, tuple(map(X_ACES, k)), 'o-', color='green', label='X - ACES RGB')

settings = {'title': 'sRGB - ACES RGB - Uniformly Spaced Incrementing RGB',
            'x_label': 'k',
            'y_label': 'X',
            'x_tighten': True,
            'legend': True,
            'legend_location': 'upper left',
            'x_ticker': True,
            'y_ticker': True}
aspect(**settings)
display(**settings)
```

They match each other very well, which is the expected and correct behavior, now we replot the functions by only incrementing the $R$ component:

```python
k = np.linspace(0, 1, 10)
X_sRGB = lambda x: 0.41238656 * x
X_ACES = lambda x: 0.9525524 * x

pylab.plot(k, tuple(map(X_sRGB, k)), 'o-', color='red', label='X - sRGB')
pylab.plot(k, tuple(map(X_ACES, k)), 'o-', color='green', label='X - ACES RGB')

settings = {'title': 'sRGB - ACES RGB - Uniformly Spaced Incrementing R',
            'x_label': 'k',
            'y_label': 'X',
            'x_tighten': True,
            'legend': True,
            'legend_location': 'upper left',
            'x_ticker': True,
            'y_ticker': True}
aspect(**settings)
display(**settings)
```

The slopes are obviously very different now and this is the reason why secondary bounces may quickly lead to very saturated colours while rendering with *ACES RGB* primaries instead of *sRGB* ones.

This can be again demonstrated by multiplying our *dark skin* and *green* colours from above computations and plotting them:

```python
pylab.plot(X_sRGB(0.172906) * X_sRGB(0.05440562), 0, 'o', color='red', label='X - "dark skin" * "green" sRGB')
pylab.plot(X_ACES(0.11758989) * X_ACES(0.15129818), 0, 'o', color='green', label='X - "dark skin" * "green" ACES RGB')

settings = {'title': 'sRGB - ACES RGB - X Value - "dark skin" * "green"',
            'x_label': 'X Value',
            'x_tighten': True,
            'legend': True,
            'legend_location': 'upper left',
            'x_ticker': True,
            'y_ticker': True}
aspect(**settings)
display(**settings)
```

## Conclusion

While most of this is really trivial linear algebra, it has been overlooked by a fair number of VFX industry veterans, including myself. We were assuming wrongly that rendering engines are colourspaces agnostic because we were told so.

On a strictly technical point of view, rendering engines are indeed colourspaces agnostic. They just chew through whatever data you throw at them without making any consideration of the colourspace that data is stored into.

However the choice of colourspace and its primaries is critical to achieve a faithful rendering. Wide gamut colour spaces are creating very colours with very high chroma values in rendering context and this can be a major issue when trying to achieve realistic output. In some instances, imaginary colours sitting outside the boundaries of the spectral locus can be generated.

The root cause of those issues is deeply linked to the current $RGB$ lighting model that *most* rendering engines use.

There are a lot of practical reasons to use the $RGB$ lighting model instead of a full spectral model but the most important one is that we don’t have an apparatus that allow us to quickly capture reflectance data on large scale nor the tools to work with that reflectance data. Nowadays, we can't ask a texture artist to launch his Spectral Photoshop or Spectral Mari and do a Spectral Gaussian Blur on his Spectral Textures.

We may be often tempted to think that textures represent true reflectance values, but they are not, they are simply factors of a given $RGB$ colourspace primaries.

To make matter worse, pretty much all of the time, photos / plates are shot with a lighting rig that doesn't match any known illuminant at best or worse, mixes multiple light sources that don't have the same relative spectral power distributions.

HDRI images used for lighting are no exception, everything is actually concerned. A light, will only act as a multiplier of the $RGB$ factors represented by the textures. This for example shortcut an important aspect of colorimetry: [metamerism](http://en.wikipedia.org/wiki/Metamerism_(color)), which is actually perfectly modeled by spectral renderers.

When we define the colour temperature of a light, we don't actually use the planckian radiatior spectral power distribution for that temperature, we just use the final colour from that spectral power distribution, probably with *sRGB* primaries, and this is assuming the planckian radiator colour has not been computed using an approximation.

Within the $RGB$ lighting model, in production context, it's theoretically impossible to compute precise colorimetric values for another viewing condition mimicking what would exactly happen in the the real world. 

For now we are just *emulating* real life, not *simulating* it. We won’t be able to do it unless we have the spectral power distributions for all our dataset (for each pixel of each images, for each light sources, etc...).

We can approximate the new viewing condition and produce something perceptually faithful and pleasing thanks to our artistic sense, we have been doing that for years and this is probably the most important thing for the VFX industry. When seeing all the current box office hits, one can say there isn't really anything we can't do in term of effects.

Ultimately, the solution will be to switch to spectral rendering, and it will happen. Question is when will it become mainstream? There are already quite a lot of spectral renderers outside, Weta has been using [Manuka](http://www.fxguide.com/featured/manuka-weta-digitals-new-renderer/) on a few shows already.
